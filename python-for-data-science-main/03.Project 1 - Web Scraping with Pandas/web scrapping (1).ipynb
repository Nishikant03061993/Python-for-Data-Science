```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
# creating list of leagues
leagues = ['E0','E2', 'E3']

frames = []
# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + "2122" + "/"  + league + ".csv")
    frames.append(df)
```

```python
import pandas as pd
import time

frames = []

for league in leagues:
    for season in range(15, 21):
        url = f"{root}{season}{season+1}/{league}.csv"
        print("Loading:", url)

        try:
            df = pd.read_csv(url)

            # avoid duplicate column error
            if 'season' not in df.columns:
                df.insert(1, 'season', season)

            frames.append(df)
            time.sleep(2)  # IMPORTANT

        except Exception as e:
            print(f"Skipped {url} -> {e}")
```

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionary element
dict_countries['German Bundesliga']
```

```python
import time
import pandas as pd

dict_historical_data = {}

for league_name, league_code in dict_countries.items():
    print(f"\nProcessing league: {league_name}")
    frames = []

    for season in range(15, 21):
        url = f"{root}{season}{season+1}/{league_code}.csv"
        print("  Loading:", url)

        try:
            df = pd.read_csv(url)

            # prevent duplicate column error
            if 'season' not in df.columns:
                df.insert(1, 'season', season)

            frames.append(df)
            time.sleep(2)  # IMPORTANT

        except Exception as e:
            print(f"  Skipped season {season}: {e}")

    if frames:  # only concat if data exists
        dict_historical_data[league_name] = pd.concat(frames, ignore_index=True)
```
